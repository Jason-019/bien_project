# 04 — 动机原型 (V4 克莱因四元群)

**BWV861 音乐形式化分析系统** · Step 4

实现 V4 = {e, I, R, RI} 群变换，计算轨道 Orbits 和稳定子 Stab，提取动机原型 μ。

关键等式:
- e·φ = (I, R̂, T̂)
- I·φ = (-I, R̂, T̂)
- R·φ = (-I^rev, R̂^rev, T̂^rev)
- RI·φ = (I^rev, R̂^rev, T̂^rev)

μ = Orb(φ) = {g·φ | g∈V4}

## 4.1 构建动机原型

```python
import sys, os
PROJECT_ROOT = r"e:\大三\Cline Projects\bien_project"
os.chdir(PROJECT_ROOT)
sys.path.insert(0, PROJECT_ROOT)

from music_analysis.features import load_features
from music_analysis.motifs import (
    build_motif_prototypes, save_motifs,
    extract_phi_rel, compute_orbit, compute_stabilizer,
)

features, single_windows, multi_windows = load_features()
print(f"✅ 已加载 {len(features)} 条特征")
```

```python
motifs = build_motif_prototypes(features)
save_motifs(motifs)

# 按出现频次排序
motifs_by_freq = sorted(motifs, key=lambda m: len(m["window_indices"]), reverse=True)
print(f"\n前 10 个最常见的动机原型:")
for m in motifs_by_freq[:10]:
    print(f"  μ#{m['motif_id']}: |Orb|={m['orbit_size']}, 出现{len(m['window_indices'])}次, "
          f"类型={m['symmetry_type']}")
```
